```python
from bs4 import BeautifulSoup
import lxml
import requests

sn = requests.Session()
url = "https://securitytracker.com/topics/topics.html"
html = requests.get(url).text
r = sn.get(url=url)
soup = BeautifulSoup(r.content)
#print(soup.prettify())
```

```python
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
```

```python

categories = []
for link in links:
    try:
        if link.startswith('/archives/impact'):
            categories.append(link)
    except:
        continue
print(len(categories))

```

```python
def get_urls_from_next_page(soup ):
    next_url = ''
    for i,page in enumerate(soup.find_all('a')):
        #print(i,page.string)
        if i>6:
            if page.string=='Previous Page':
                if soup.find_all('a')[i+1].string!='Next Page':
                    break
                continue
            if page.string=='Next Page':
                next_url = page.get('href')
                break
            else:
                # print(page.get('href'))
                page_urls.append(page.get('href'))
                
    return next_url, page_urls

report_urls = []

for cat in categories:
    #print(cat)
    url = "https://securitytracker.com"+cat
    html = requests.get(url).text
    r = sn.get(url=url)
    soup = BeautifulSoup(r.content)
    print('get next page and report links from url:',url)
    next_page_url,page_report_urls = get_urls_from_next_page(soup)
    report_urls.extend(page_report_urls)
    while next_page_url:
        url = "https://securitytracker.com/archives/impact/"+next_page_url
        #print('page',url)
        html = requests.get(url).text
        r = sn.get(url = url)
        soup = BeautifulSoup(r.content)
        print('get next page and report links from url:',url)
        next_page_url,page_report_urls = get_urls_from_next_page(soup)
        #print('next page',next_page_url)
        report_urls.extend(page_report_urls)
        
print(len(report_urls))
```

```python
import os  
import time
def mkdir(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
RECORD_URL_LIST = 0

resultpath = '../../data/securityTracker/report/'
mkdir(resultpath)

result_file = []
for filename in os.listdir(resultpath):
    result_file.append('/id/'+filename.replace('.txt',''))

print("crawled documents:",len(result_file))
if RECORD_URL_LIST:
    count = 0
    with open('../../data/securityTracker/report_list.txt','w') as f:
        for rid in report_urls:
            if str(rid).startswith('/'):
                f.write(str(rid))
                f.write('\n')
                count+= 1
    print(count)



report_urls = list(set(report_urls))
print(len(report_urls))
for iid,rid in enumerate(report_urls):
    rid = str(rid)
    if rid in result_file or not rid.startswith('/'):
        #print('crawled',rid)
        continue
    if iid%50 ==0:
        print(iid,rid)
       
    try:
        url = "https://securitytracker.com"+rid
        #print('page',url)
        html = requests.get(url).text
        r = sn.get(url = url)
        soup = BeautifulSoup(r.content)

        links = soup.find_all('tr')
    except Exception as e:
        continue
    # intialize
    vendors = ''
    product = ''
    cve_list = []
    impact = []
    intrusions = []
    fix_available = 0
    vendor_confirmed = 0
    researcher = ''
    affected_version = []
    fix = ''
    for link in links:
        text = link.text.replace('\n\n','\n').replace('\xa0','').split('|')
        #print(i,text)
        if 'Vendors:' in text[0] and 'Category:' in text[0]:
            info = text[0].split('\n')
            #print(info)
            for i,item in enumerate(info):
                if item =='Category:':
                    product = info[i+2]
                if item=='Vendors:':
                    vendor = info[i+1]
                    
                if vendor in product:
                    product = product.replace(vendor,'').strip(' ')
                    
                if item == 'CVE Reference:':
                    cve_list = info[i+1].split(',')
                if 'Impact:' in item:
                    if impact == []:
                        impact = info[i+1].split(',')
                if 'Fix Available:' in item:
                    if item[item.index('Fix')+len('Fix Available:')]=='Y':
                         fix_available = 1
                if 'Vendor Confirmed:' in item:
                    if item[item.index('Vendor')+len('Vendor Confirmed:')]=='Y':
                         vendor_confirmed = 1
                if ' [CVE' in item:
                    intrusions.append(item)
                if 'reported' and 'vulnerabilities' in item:
                    researcher = item
                if 'Version(s):' in item:
                    affected_versions = item
                if 'Solution:' in item:
                    fix =info[i+1]
                if 'Underlying OS' in item:
                    break
                    
                    
    with open(resultpath+rid.replace('/id/','')+'.txt','w') as f:
        f.write('-cve: ')
        for cve in cve_list:
            f.write(cve+'|')
        f.write('\n\n-product: ' + product )
        f.write('\n\n-vendor: ' + vendor)
        f.write('\n\n-version: '+ affected_versions)
        f.write('\n\n-solution: '+fix)
        f.write('\n\n-impact: ')
        for im in impact:
            f.write(im + '|')
        f.write('\n\n-vendor confirmed: '+str(vendor_confirmed))
        f.write('\n\n-fix available: '+str(fix_available))
        f.write('\n\n-intrusions: ')
        for intrusion in intrusions:
            f.write(intrusion+'|')
        f.write('\n\n-researcher: '+researcher)
        #print("{}-th file, finish writing:{}".format(iid,resultpath+rid.replace('/id/','')+'.txt'))

           
            
        
```
