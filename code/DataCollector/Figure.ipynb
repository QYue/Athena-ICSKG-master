```python
import numpy as np

import pandas as pd

import matplotlib as mpl

import matplotlib.pyplot as plt

import seaborn as sns

import warnings; warnings.filterwarnings(action='once')



large = 22; med = 16; small = 12

params = {'axes.titlesize': large,

          'legend.fontsize': med,

          'figure.figsize': (16, 10),

          'axes.labelsize': med,

          'axes.titlesize': med,

          'xtick.labelsize': med,

          'ytick.labelsize': med,

          'figure.titlesize': large}

plt.rcParams.update(params)

plt.style.use('seaborn-whitegrid')
plt.xlim((0,10))
plt.ylim((0,10))
plt.xticks(np.arange(0, 10, 1))
plt.yticks(np.arange(0, 10, 1))
sns.set_style("white")

%matplotlib inline
# Import Data

#df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/mpg_ggplot2.csv")

df=pd.read_csv("../../../result/cvss_change.csv")
df_counts = df.groupby(['old', 'new']).size().reset_index(name='counts')



# Draw Stripplot

fig, ax = plt.subplots(figsize=(12,7.5), dpi= 80)    

sns.stripplot(df_counts.old, df_counts.new, size=df_counts.counts*3, ax=ax)


plt.tick_params(labelsize=10)
ax.set_yticks(np.arange(0, 10,1))  


# Decorations

plt.title('Size of circle is bigger as more points overlap', fontsize=22)

plt.show()

fig.savefig('../../Figures/cvss_change.jpg')
```
